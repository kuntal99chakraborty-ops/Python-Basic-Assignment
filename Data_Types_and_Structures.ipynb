2. Explain the difference between mutable and immutable data types with examples.

- Difference between mutable and immutable assets are

Mutable Assets

Mutable data types are those whose values can be modified in place after they are created. This means that operations performed on a mutable object can change its internal state without creating a new object in memory.

Examples: Lists, Dictionaries, Sets in Python.

Immutable assets

Immutable data types are those whose values cannot be changed after they are created. Any operation that appears to modify an immutable object actually results in the creation of a new object with the desired changes, leaving the original object unchanged.

Examples: Numbers (integers, floats), Strings, Tuples in Python.

8. What is a hash table, and how does it relate to dictionaries in Python?

- A hash table is a data structure that implements an associative array, also known as a dictionary or map. It stores key-value pairs and allows for efficient retrieval of values based on their associated keys.

 Examples through which it relate to dictionaries in Python

 * Hashing: When you want to store a key-value pair, the hash function takes the key as input and produces an integer hash value.
 Indexing: This hash value is then typically used with the modulo operator (%) and the size of the underlying array to determine the index where the value should be stored.

* Storage: The value is placed in the array at the calculated index. If multiple keys hash to the same index (a "collision"), various techniques like separate chaining (using linked lists at each index) or open addressing (probing for the next available slot) are used to handle them.

14. How does the “in” keyword work differently for lists and dictionaries?

- The in keyword in Python functions differently for lists and dictionaries due to their underlying data structures and how they store and retrieve data.

For Lists:

* When used with a list, the in operator checks for the presence of a specific element within the list's values.

* It performs a linear search, iterating through each element in the list until a match is found or the end of the list is reached.

* The time complexity for this operation is typically O(n) in the worst case, meaning it can take longer as the list grows.

For Dictionaries:

* When used with a dictionary, the in operator checks for the presence of a specific key within the dictionary.
* It does not check for the presence of values directly.
* Dictionaries are implemented using hash tables, which allow for very efficient lookups based on keys.
* The time complexity for checking key existence in a dictionary is typically O(1) on average, making it much faster than searching in a list for large collections.

17. Describe the time complexity of accessing elements in a dictionary.

- The time complexity of accessing elements in a dictionary, such as in Python, is generally O(1) on average. This means that the time taken to retrieve a value associated with a key is constant, regardless of the number of elements in the dictionary.

This efficiency stems from the underlying implementation of dictionaries as hash tables. When a key is used to access an element:

* Hashing: The key is passed through a hash function, which converts it into an integer hash value.

* Indexing: This hash value is then used to determine the memory location (or "bucket") where the corresponding value is stored.

18. In what situations are lists preferred over dictionaries?

- Lists are preferred over dictionaries in several situations, primarily when the order of elements is crucial or when data is accessed sequentially.

* Ordered Collections: When the order of elements matters, such as a sequence of steps, a queue, or a history of events, lists are the appropriate choice. Dictionaries, while maintaining insertion order in modern Python versions, are fundamentally designed for key-value associations, not ordered sequences.

* Sequential Processing: For operations that involve iterating through elements in a specific order, like applying a function to each item in a sequence or performing calculations based on the position of elements, lists are more intuitive and efficient.

* Simple Data Storage: For collections of data where each element is self-contained and does not require a unique identifier for retrieval, lists offer a simpler and often more memory-efficient way to store and manage the data.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?

- Dictionaries are considered "unordered" because, historically and by design, their primary purpose is fast, key-based access, not maintaining the order of insertion or any other sequence, though Python 3.7+ preserves insertion order as an implementation detail. This unordered nature means you cannot rely on a specific sequence when iterating or printing items, and data retrieval is always based on the unique key, not a positional index.

How it affects data retrieval?

- Access by Key: Data is always retrieved using its unique key, not its position. For example, to get a value, you must know the corresponding key, such as my_dictionary['apple'].

No Order Guarantees (Conceptually): Even though modern Python dictionaries maintain insertion order, you should not depend on this for the overall function of the data structure. When iterating through a dictionary, the order of elements may not be predictable or match the insertion order in contexts where insertion order isn't relevant.

Fast Lookups: The primary benefit of this design is speed. Hashing (using a function to convert the key into an index) allows for very fast retrieval of values associated with a given key.

20. Explain the difference between a list and a dictionary in terms of data retrieval.

Difference between lists and dictionaries in terms of data retrieval

* Lists:

* Ordered Collection: Lists maintain a specific order of elements, meaning the position of each element is fixed relative to the others.

* Indexed Retrieval: Data retrieval in lists is primarily done using integer indices, starting from 0 for the first element.

Dictionaries:

* Unordered Key-Value Pairs: Dictionaries store data as key-value pairs, where each unique key is associated with a specific value. The order of these pairs is not guaranteed.

* Key-Based Retrieval: Data retrieval in dictionaries is performed using the associated key, which can be of various immutable data types (e.g., strings, numbers, tuples).
