9) What is type casting in Python?

- Type casting, also known as type conversion. Type casting is the process of converting a variable's data type into another data type.

10) What is the difference between implicit and explicit type casting?

- Difference between implicit and explicit type casting are:

Implicit type casting
* It is automatic as the coversion is performed automatically by the compiler or interpreter without programmer intervention.
* It occurs when a value from a smaller data type is converted to a larger data type.
* This process is safe as the larger type can hold the entire value of the smaller type.

Explicit type casting
*  It is manual and the programmer must explicitly write the code to perform the conversion.
* It involves converting a value from a larger data type to a smaller one.
* This can lead to data loss or truncation because the smaller type may not be able to accommodate the full range of values from the larger type.

13) What is the difference between for and while loops?

*  for loops
- for loop is used for a known, fixed number of iterations.
- for loops have a concise structure for initialization, condition, and increment within the loop declaration, making them ideal for iterating over collections or a predetermined range.

*  while loops

- while loop is used when the number of iterations is unknown and depends on a dynamic condition.
- while loops check a condition before each iteration and require the loop body to manage any necessary updates to keep the condition from becoming permanently true, which could lead to an infinite loop.

14) Describe a scenario where a while loop is more suitable than a for loop.

- A while loop is more suitable than a for loop when the number of iterations is unknown or unpredictable in advance, such as waiting for user input until a valid entry is provided, reading data from a file until the end is reached, or streaming data from a database until a specific condition is met. In these cases, a for loop is less ideal because it typically iterates a predefined number of times.

```python
6. Write a function to demonstrate the behavior of mutable and immutable arguments.
```

```python
def modify_mutable_arg(my_list):
    """
    Demonstrates how modifying a mutable argument affects the original object.
    """
    print(f"Inside function (mutable): Initial ID: {id(my_list)}, Value: {my_list}")

    # Modify the list in-place
    my_list.append(4)
    print(f"Inside function (mutable): Modified ID: {id(my_list)}, Value: {my_list}")

def modify_immutable_arg(my_int):
    """
    Demonstrates how modifying an immutable argument does not affect the original object.
    """
    print(f"Inside function (immutable): Initial ID: {id(my_int)}, Value: {my_int}")

    # Reassign the variable to a new value
    my_int += 1
    print(f"Inside function (immutable): Modified ID: {id(my_int)}, Value: {my_int}")

# --- Demonstration with a mutable argument (list) ---
print("--- Demonstrating mutable arguments (lists) ---")
original_list = [1, 2, 3]
print(f"Outside function: Original ID: {id(original_list)}, Value: {original_list}")
modify_mutable_arg(original_list)
print(f"Outside function: Final ID: {id(original_list)}, Value: {original_list}")

# --- Demonstration with an immutable argument (integer) ---
print("\n--- Demonstrating immutable arguments (integers) ---")
original_int = 10
print(f"Outside function: Original ID: {id(original_int)}, Value: {original_int}")
modify_immutable_arg(original_int)
print(f"Outside function: Final ID: {id(original_int)}, Value: {original_int}")
```

7. Write a program that performs basic arithmetic operations on two user-input numbers.

```python
# Write a Python program that performs basic arithmetic operations on two user-input numbers.

# 1. Get user input for two numbers
# Use float() to allow for decimal numbers
try:
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
except ValueError:
    print("Invalid input. Please enter numerical values.")
else:
    # 2. Perform arithmetic operations
    addition = num1 + num2
    subtraction = num1 - num2
    multiplication = num1 * num2

    # Division requires special handling to prevent an error if dividing by zero
    if num2 != 0:
        division = num1 / num2
    else:
        division = "Cannot divide by zero"

    # Floor division (integer division)
    if num2 != 0:
        floor_division = num1 // num2
    else:
        floor_division = "Cannot perform floor division by zero"

    # Modulus (remainder)
    if num2 != 0:
        modulus = num1 % num2
    else:
        modulus = "Cannot calculate modulus with zero"

    # Exponentiation (power)
    exponentiation = num1 ** num2

    # 3. Print the results using f-strings for clear formatting
    print("\n--- Results ---")
    print(f"Addition: {num1} + {num2} = {addition}")
    print(f"Subtraction: {num1} - {num2} = {subtraction}")
    print(f"Multiplication: {num1} * {num2} = {multiplication}")
    print(f"Division: {num1} / {num2} = {division}")
    print(f"Floor Division: {num1} // {num2} = {floor_division}")
    print(f"Modulus: {num1} % {num2} = {modulus}")
    print(f"Exponentiation: {num1} ** {num2} = {exponentiation}")
```

8. Write a program to demonstrate the use of logical operators.

```python
# Assume these variables represent certain conditions
is_logged_in = True
is_admin = False
has_license = True
is_underage = False

print("--- Demonstrating the 'and' operator ---")
# The 'and' operator returns True only if both operands are True.
if is_logged_in and has_license:
    print("User is logged in and has a license.")
else:
    print("User does not meet both conditions.")

print("\n--- Demonstrating the 'or' operator ---")
# The 'or' operator returns True if at least one of the operands is True.
if is_admin or is_underage:
    print("User is either an admin or is underage.")
else:
    print("User is neither an admin nor underage.")

print("\n--- Demonstrating the 'not' operator ---")
# The 'not' operator reverses the boolean value of the operand.
if not is_admin:
    print("User is NOT an admin.")
else:
    print("User is an admin.")

print("\n--- Combining logical operators ---")
# Logical operators can be combined for more complex conditions.
if is_logged_in and has_license and not is_underage:
    print("The user is a logged-in adult with a license.")
else:
    print("The user does not meet all criteria.")
```

9. Write a Python program to convert user input from string to integer, float, and boolean types.

```python
import sys

def get_int_input():
    """Converts user input to an integer with error handling."""
    while True:
        user_input = input("Enter an integer: ")
        try:
            return int(user_input)
        except ValueError:
            print(f"Invalid input: '{user_input}' cannot be converted to an integer. Please try again.")

def get_float_input():
    """Converts user input to a float with error handling."""
    while True:
        user_input = input("Enter a float: ")
        try:
            return float(user_input)
        except ValueError:
            print(f"Invalid input: '{user_input}' cannot be converted to a float. Please try again.")

def get_boolean_input():
    """
    Converts user input to a boolean.
    Accepts "true", "t", "yes", or "y" (case-insensitive) as True.
    Accepts "false", "f", "no", or "n" (case-insensitive) as False.
    All other non-empty strings are treated as False, with a warning.
    An empty string is also False.
    """
    true_options = {"true", "t", "yes", "y"}
    false_options = {"false", "f", "no", "n"}

    while True:
        user_input = input("Enter a boolean (e.g., True/False, Y/N): ")

        # Strip whitespace and convert to lowercase for case-insensitive comparison
        cleaned_input = user_input.strip().lower()

        if cleaned_input in true_options:
            return True
        elif cleaned_input in false_options:
            return False
        elif cleaned_input == "":
            print("Empty input. Interpreted as False.")
            return False
        else:
            print(f"Unknown boolean value: '{user_input}'. Interpreted as False.")
            return False

def main():
    """
    Main function to run the program.
    """
    print("--- User Input Type Conversion Program ---")

    # 1. Convert to Integer
    print("\n--- Testing Integer Conversion ---")
    my_integer = get_int_input()
    print(f"You entered the integer: {my_integer} (Type: {type(my_integer)})")

    # 2. Convert to Float
    print("\n--- Testing Float Conversion ---")
    my_float = get_float_input()
    print(f"You entered the float: {my_float} (Type: {type(my_float)})")

    # 3. Convert to Boolean
    print("\n--- Testing Boolean Conversion ---")
    my_boolean = get_boolean_input()
    print(f"You entered the boolean: {my_boolean} (Type: {type(my_boolean)})")

if __name__ == "__main__":
    main()
```

13. Write a Python program to find the sum of all even numbers between 1 and 50.

```python
# Initialize a variable to store the sum
even_sum = 0

# Loop through numbers from 1 up to 50 (51 is exclusive in range)
for number in range(1, 51):
  # Check if the number is even
  if number % 2 == 0:
    # If it is, add it to the sum
    even_sum += number

# Print the final result
print(f"The sum of all even numbers between 1 and 50 is: {even_sum}")
```

14.  Write a program to reverse a string using a while loop.

```python
def reverse_string_while_loop(input_string):
  """
  Reverses a string using a while loop by iterating backward.
  """
  # Initialize an empty string to build the reversed version
  reversed_string = ""

  # Set the index to the last character of the input string
  index = len(input_string) - 1

  # Loop until the index is less than 0 (i.e., we have gone through all characters)
  while index >= 0:
    # Add the character at the current index to the new string
    reversed_string += input_string[index]
    # Decrement the index to move to the previous character
    index -= 1

  return reversed_string

# Example usage
original_str = "hello world"
reversed_str = reverse_string_while_loop(original_str)

print(f"Original string: {original_str}")
print(f"Reversed string: {reversed_str}")
```

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
# Get user input for the number
try:
    num = int(input("Enter a non-negative integer: "))

    # Handle negative numbers, zero, and positive numbers
    if num < 0:
        print("Factorial does not exist for negative numbers.")
    elif num == 0:
        print("The factorial of 0 is 1.")
    else:
        # Initialize factorial and a counter
        factorial = 1
        i = 1

        # Use a while loop to calculate the factorial
        while i <= num:
            factorial *= i
            i += 1

        print(f"The factorial of {num} is {factorial}.")

except ValueError:
    print("Invalid input. Please enter an integer.")
```
